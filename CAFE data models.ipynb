## This notebook shows the datamodel for the CAFE project

```python
from flighttools import flight, visuals
import openlattice
import os

basedir = "/Users/jokedurnez/Documents/Integrations/CAFE-integrations/"
```

```python
baseurl = 'http://localhost:8080'
# baseurl = 'https://api.openlattice.com'
jwt = "*"

configuration = openlattice.Configuration()
configuration.host = baseurl
configuration.api_key_prefix['Authorization'] = 'Bearer'
configuration.api_key['Authorization'] = jwt

edmAPI = openlattice.EdmApi(openlattice.ApiClient(configuration))
```

# Time Use Diary

```python
filename= os.path.join(basedir,'time_use_diary_template.yaml')
fl = flight.flight(edmAPI)
fl.deserialise(filename)
fl.add_and_check_edm()
```

```python
visualisation = visuals.edmViz(fl)
fullplot = visualisation.create_flight_plot(type="entityset_name")
fullplot.format = 'png'
fullplot
```

```python
visualisation = visuals.edmViz(fl, engine = 'circo')
fullplot = visualisation.create_flight_plot(
    type="name", split_associations=False, plotlinks=False, fields=['properties'])
fullplot
```
