## This notebook shows the datamodel for the CAFE project

```python
from flighttools import flight, visuals
import openapi_client
import os

basedir = "/Users/jokedurnez/Documents/Integrations/CAFE-integrations/"
```

```python
baseurl = 'http://localhost:8080'
jwt = "*"

configuration = openapi_client.Configuration()
configuration.host = baseurl
configuration.api_key_prefix['Authorization'] = 'Bearer'
configuration.api_key['Authorization'] = jwt

edmAPI = openapi_client.EdmApi(openapi_client.ApiClient(configuration))
```

```python
filename= os.path.join(basedir,'time_use_diary.yaml')
fl = flight.flight(edmAPI)
fl.deserialise(filename)
fl.add_and_check_edm()
```

```python
visualisation = visuals.edmViz(fl)
fullplot = visualisation.create_flight_plot(type="fqn")
fullplot.format = 'png'
# fullplot.render(os.path.join(imdir, "Offenders_small"))
fullplot
```

```python
visualisation = visuals.edmViz(fl)
fullplot = visualisation.create_flight_plot(
    type="entityset_name", split_associations=False, plotlinks=True, fields = ['properties'])
# fullplot.render(os.path.join(basedir, "Model_name"))
fullplot
```
